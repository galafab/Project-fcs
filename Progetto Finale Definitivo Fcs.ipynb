# Esercizio 9
For each (route, bus) pair, compute the ratio between the overall energy consumption and the overall driven distance.

```python
gruppi=bus.groupby(['busRoute','busNumber']).sum()
energyratio=gruppi['energyConsumption']/gruppi['drivenDistance']
energyratio
```

# Esercizio 10
Starting from the results of the previous point, for each route compute the buses with max and min energy ratio, and save the difference between these ratios in a dataframe.

```python
enratio=energyratio.reset_index()
#gli indici delle osservazioni vengono resettaati, la terza colonna assume nome '0' perciò lo cambio col seguente comando
enratio=enratio.rename(columns={0:'range'})
es10=pd.DataFrame((enratio.groupby(['busRoute']).max()-enratio.groupby(['busRoute']).min()))
es10=es10.reset_index()

es10=es10.iloc[:,[0,2]]
print(enratio)
es10
```

# Esercizio 11
Find the bus maximizing the difference computed in the previous point.

```python
bestRoute=es10.iloc[es10['range'].idxmax()][0]  #busRoute of the best
bestbuses=enratio[enratio['busRoute']==bestRoute] #the best buses
bestbusindex=bestbuses['range'].idxmax() #index of the best bus
bestbus=bestbuses.loc[bestbusindex]
bestbus
#as you can see the best is number 183 for route 31
```

```python
#add the trip that change month, one observation for each month reporting the average temperature
for i in cambiamese:
    percorso_file = os.path.join(cartella_dataset, elenco_file[i])
    bus1=pd.read_csv(percorso_file)
    bus1.time_iso=bus1.time_iso.apply(lambda x: parser.isoparse(x)).dt.month
    #qui calcolo la media della temperatura nel mese corrispondente in cui è avvenuto il trip
    a=pd.DataFrame(bus1.groupby('time_iso').mean()['temperature_ambient']).reset_index()
    a.columns=['startTime_iso','temperature_ambient_mean']
    #aggiungo queste osservazioni al dataset con busmediuguali
    busmesiuguali=pd.concat([busmesiuguali,a])
    
```

```python
stazioni=pd.read_csv("C:/Users/fabio/Desktop/Found of cs/finale.csv", low_memory=False)
stazioni=stazioni.rename(columns={'nome_bus':'trip_name'})
```

```python
trips=np.unique(stazioni.trip_name)
ultimi=[]
#salva l'indice corrispondente all'ultima osservazione di ogni trip
for i in range(0,len(trips)):
    ultimi=ultimi+[stazioni[stazioni.trip_name==trips[i]].index[-1]]
ultimi
```

```python
#creac un vettore che parte dalla prima osservazione e arriva alla fine contenente itcs_stopname
stazdopo=stazioni.itcs_stopName[range(1,len(stazioni))]
#aggiungo un'osservazione alla fine contenente '-'
stazdopo=pd.concat([stazdopo,pd.Series(['-'])], ignore_index=True)
#assegna '-' alle osservazioni contenute in ultimi
stazdopo[ultimi]='-'
stazioni['stazdopo']=stazdopo
stazioni.groupby(['itcs_stopName','stazdopo']).mean()['odometry_vehicleSpeed']
#dovrei ripetere la procedura utilizzata per creare stazdopo per riportare le variabili utili per fare il calcolo, tuttavia
#non trovo  che le variabili siano utili
```
